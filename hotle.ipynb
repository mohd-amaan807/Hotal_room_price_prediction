```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
x=df[['city','distance to city centre(km)',
       'distance to HYD-Rajiv Gandhi Intl(km)'
        , 'Pool', 'Free parking',
       'Airport transfer', 'Spa', 'Restaurant', 'Gym', 'Bar', 'Bathtub',
       'Meeting Facilities', 'Pet-friendly', 'Kitchen', 'Internet access']]
y=df["price"]
```

```python
from  sklearn.model_selection  import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=302)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
ohe=OneHotEncoder()
ohe.fit(x[['city'
        , 'Pool', 'Free parking',
       'Airport transfer', 'Spa', 'Restaurant', 'Gym', 'Bar', 'Bathtub',
       'Meeting Facilities', 'Pet-friendly', 'Kitchen', 'Internet access']])
```

```python
column_trans=make_column_transformer(
    (OneHotEncoder(categories=ohe.categories_),['city'
        , 'Pool', 'Free parking',
       'Airport transfer', 'Spa', 'Restaurant', 'Gym', 'Bar', 'Bathtub',
       'Meeting Facilities', 'Pet-friendly', 'Kitchen', 'Internet access']),
    remainder='passthrough')
```

```python
lr=LinearRegression()
pipe=make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```
